Q4:
(a) Write a function named img_to_array(path) that reads an image from a specified path and
save it as text file on local machine? (Note: use separate cases for RGB and Grey Scale
images)
(b) Load the saved file into jupyter notebook?

```python
import cv2
import os
from PIL import Image
def img_to_array(path):
    img = cv2.imread(path, cv2.IMREAD_UNCHANGED)

    if img is None:
        print("Image not found")
        return

    filename = os.path.splitext(path)[0]

    if len(img.shape) == 2:
        np.savetxt(filename + "_grayscale.txt", img)
        print("Grayscale image saved as text file")

    elif len(img.shape) == 3:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        np.savetxt(filename + "_R.txt", img[:, :, 0])
        np.savetxt(filename + "_G.txt", img[:, :, 1])
        np.savetxt(filename + "_B.txt", img[:, :, 2])
        print("RGB image saved as text files")
    else:
        print("Unsupported image format")

    #function call
img_to_array("toread.png")

```
